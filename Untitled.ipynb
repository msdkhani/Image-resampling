```python
import numpy as np
import matplotlib.pyplot as plt
import re
import cv2
```

#### Read_pgm Function help us to read Images with PGM format and set them in an array.

```python


def read_pgm(filename, byteorder='>'):

    with open(filename, 'rb') as f:
        buffer = f.read()
    try:
        header, width, height, maxval = re.search(
            b"(^P5\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n]\s)*)", buffer).groups()
    except AttributeError:
        raise ValueError("Not a raw PGM file: '%s'" % filename)
    return np.frombuffer(buffer,
                            dtype='u1' if int(maxval) < 256 else byteorder+'u2',
                            count=int(width)*int(height),
                            offset=len(header)
                            ).reshape((int(height), int(width)))
```

#### First we use read_pgm function to pur out image array into variable 'data'
##### as we can see we have an 2D array with rage of (0-255) where 0 is black pixels and 255 is white image

```python
data = read_pgm('img.pgm',byteorder='<')
plt.imshow(data, plt.cm.gray)
plt.show()
data
```

##### we get our dimension of a out image which is height out width, we want to resample it to T times of image, so we create an empty array with T * width and T * height 

```python
img_height ,img_width = data.shape[:2]
T = 2
new_width = int(img_width*T)
new_height = int(img_height*T)
resize_img = np.empty([new_height, new_width])
img_height, img_width, new_height,new_width
```

##### x,y is not the the integer number (they are float) so we should find ceil and floor number of them.

```python
def ceil(x):
    return int(x) + int((x>0) and (x - int(x)) > 0)

def floor(x):
    return int(x//1)
```

#### WE need 4 point to calculate the pixel in resized image so we should find I00,I01,I10,I11 and use 
$$ pixel(i,j) = x * y * I11 + x * (1 - y) * I10 + (1-x) * y * I01 + (1 - x) * (1 - y) * I00 $$

```python
x_r = float(img_width-1) / (new_width-1) 
y_r = float(img_height-1) / (new_height-1) 

for i in range(new_height):
    for j in range(new_width):
        
        x_l = floor(x_r * j)
        y_l = floor(y_r * i)
        x_h = ceil(x_r * j)
        y_h = ceil(y_r * i)

        x = (x_r * j) - x_l
        y = (y_r * i) - y_l
        
        #neighborhood of original image
        I00 = data[y_l, x_l]
        I10 = data[y_l, x_h]
        I01 = data[y_h, x_l]
        I11 = data[y_h, x_h]
        
        result = x * y * I11 + x * (1 - y) * I10 + (1-x) * y * I01 + (1 - x) * (1 - y) * I00
        resize_img[i][j] = result
        
```

```python
plt.imshow(resize_img, plt.cm.gray)
plt.show()
```

```python
cv2.imwrite("T={}.png".format(T), resize_img)
```
